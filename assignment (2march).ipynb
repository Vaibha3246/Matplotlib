Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

profit = np.array([3000, 8000, 1000, 10000])




<!-- answer -->

A bar plot, also known as a bar chart, is a graphical representation of categorical data with rectangular bars of lengths proportional to the values they represent. The bars can be plotted either horizontally or vertically. Each bar represents a category, and the height (or length) of the bar corresponds to the value of that category.


<!-- used for  -->

sumarizing data
communication

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit,color='grey')

plt.xlabel('profit')
plt.ylabel('company name')
plt.title('profit vs company')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)




<!-- answer -->

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data through quartiles. It displays the five-number summary of a dataset: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The box plot is particularly useful for identifying outliers and comparing distributions between different groups of data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis
fig, ax = plt.subplots()

# Create box plot
ax.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Add title and labels
ax.set_title('Box Plot of Box 1 and Box 2')
ax.set_xlabel('Box')
ax.set_ylabel('Values')

# Show plot
plt.show()
```
